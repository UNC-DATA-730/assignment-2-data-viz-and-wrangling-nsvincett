### Histogram

Make a histogram of penguin body mass. 
- Facet your chart by the `sex` column using the `facet_wrap` function and setting the `ncol` parameter to `1`.
- In your `geom_histogram` layer, set the `position` argument to `"identity"` and set `alpha`to a value between `0` and `1`.
- Adjust the `fill` scale palette similarly to how we adjusted the `color` scale palette above. I.e. change to the [ColorBrewer](https://colorbrewer2.org/) `"Set2"` palette.

```r
penguins |>
ggplot(aes(x = body_mass_g, fill = species)) + 
    geom_histogram(na.rm = T,position = "identity", alpha = 0.5) +
    facet_wrap(~sex, ncol = 1) +
    scale_fill_brewer(palette="Set2")
```

What happens when you don't have `position = "identity"` in your `geom_histogram` layer? Answer below in a `Markdown` cell.

The 'position = "identity"' argument tells the histogram to overlay the chart categories (in this case 'Island') on top of each other to give the appearance of three separate bars in a given space. 

The default position of geom_histogram (which occurs when no position is deliberately set) is "stack" which places the categories on top of each other creating a single continuous bar segmented into the various categories of the chart.

## Data wrangling

First we are going to bring in a data set to practice with called `txhousing`. You can find some information about `txhousing` [here](https://ggplot2.tidyverse.org/reference/txhousing.html).

```r
# uncomment the next two lines and execute this cell
 data(txhousing)
 glimpse(txhousing)
```

Write a code chunk to remove the inventory variable. Save the results in a data frame called `txhousing`. Use `glimpse` or `head` to confirm you have removed the variable.

```r
txhousing <- txhousing |>
    select(-inventory) |>
    glimpse()
```

Make a data set called `dallas_sub` that includes data only from the city of Dallas in 2012 & 2013.

```r
dallas_sub <- txhousing |>
    filter(year == 2012 | year == 2013, city == "Dallas") |>
    glimpse()
```

Add a column to the `dallas_sub` data set called `prct_sold` that calculates the percentage of listings that were sold (`sales/listings * 100`). Be sure to save the results also as a data frame called `dallas_sub`.

```r
dallas_sub <- dallas_sub |>
    mutate(prct_sold = sales/listings * 100) |>
    glimpse()
```

Calculate the average percentage of listings that were sold in Dallas in each month of the year based on your `dallas_sub` data set. Save the results of the calculation in an data frame called `dallas_summary`.

```r
dallas_summary <- dallas_sub |>
    group_by(month) |>
    summarize(average_prct_sold = mean(prct_sold, na.rm=TRUE)) |>
    glimpse()
```

Use `slice_max` to see the top 3 months with the greatest average percentage of listings sold from `dallas_summary`. You do not need to save the results.

```r
dallas_summary |>
    slice_max(average_prct_sold,n = 3)
```

Run the following code chunk. Study the code, and the output. Explain in your own words what this code chunk calculated.

```r
txhousing |>
   filter(year == 2012 | year == 2013, city == "Dallas") %>%
   mutate(prct_sold = sales/listings *100) |>
   group_by(month) |>
   summarize(mean_prct_sold = mean(prct_sold)) |>
   slice_max(mean_prct_sold, n = 3)
```

This code outputs the top three months with the most percentage of listings sold; in other words it combines and runs the previous four code snippets all in one execution (with the major difference that dataframes are not saved along the way, which does save time and readability but does not allow the sub-dataframes to be accessed at a later point or for additional executions so there are benefits and downsides to this approach).

txhousing is passed to a filter function which pulls the entries of Years 2012 or 2013, AND in the city of Dallas. 
The resulting dataframe is then passed to a mutate function which adds the prct_sold column (using the included formula) to the dataframe.
The result is passed to a group_by function which then splits rows to only associate with rows of the same 'month' value, future operations will now feed into these discrete groups instead of the dataframe as a whole.
This is passed to a summarize function, which averages the prct_sold value of each row and returns the resulting value; because we grouped rows by the 'month' column it is included in the summary, there will be a separate row for each month.
Finally this is passed to the slice_max function, which takes a column to identify and return the 'n' rows with the highest value in that column. Here we asked for mean_prct_sold and n=3, and so we receive the three months with the highest mean_prct_sold values.
